```python
import datasets
import pandas as pd
```

```python
def open_file(file_name):
    res = []
    with open(file_name, 'r') as f:
        for line in f.readlines():
            res.append(line.strip())
    return res

en_data = open_file('en.txt')
vi_data = open_file('vi.txt')
zh_data = open_file('zh.txt')
    
    
```

```python
df_en = pd.DataFrame({'text':en_data, 'lang':['en' for _ in range(len(en_data))]})
df_vi = pd.DataFrame({'text':vi_data, 'lang':['vi' for _ in range(len(vi_data))]})
df_zh = pd.DataFrame({'text':zh_data, 'lang':['zh' for _ in range(len(zh_data))]})
```

```python
df = pd.concat([df_en, df_vi, df_zh])
```

```python
ds = datasets.Dataset.from_pandas(df)
```

```python
df.sample(5)
```

```python
ds.to_json('all_data.json', force_ascii=False)
```
